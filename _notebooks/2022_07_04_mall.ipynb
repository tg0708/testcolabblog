```python
from tqdm import tqdm
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd


col = ['날짜', '카테고리', '홈사','시간','타이틀','제품가','이미지']

df = pd.DataFrame(columns=col)

date = input('날짜를 입력하세요 : ')
res = requests.get(f'https://www.hsmoa.com/?date={date}&site=&cate=')
res.status_code
soup = BeautifulSoup(res.content, "html.parser")

item_list = []
category_list = []
shopmall_list = []
time_list = []
price_list = []

container = soup.select('div.timeline-item')
for idx, item in tqdm(enumerate(container)):
    # 상품
    item_content = item.select_one('.font-15').text.strip()
    item_list.append(item_content)
    # 문자화    
    make_str = str(item)
    make_str = make_str.split('>')[0]

    # 쇼핑사
    shopmall_content = make_str.split('timeline-item ')[1].split(" ")[0]
    shopmall_list.append(shopmall_content)

    # 카테고리 분류
    p = re.compile('[가-힣]')
    result = p.findall(make_str)
    category_content = "".join(result)
    category_list.append(category_content)

    # 시간0
    try :
        time_contents = item.select_one('.font-12').text.strip()
        time_start = time_contents.split("\n")[0]
        time_end = time_contents.split("\n")[1].strip()
        time_content = time_start + time_end
        time_list.append(time_content)
    except:
        time_content = "-"


    # 가격
    price_contents = item.select_one('.strong.font-17').text.strip()

    # 이미지
    str_img = str(item.select_one('img.lazy'))
    img_content = str_img.split('smart/')[1].split('"')[0]

    # 딕셔너리
    dict_info = {'날짜':date, '카테고리': category_content,'홈사':shopmall_content,
             '시간':time_content,'타이틀':item_content,'제품가':price_contents,'이미지':img_content}

    df = df.append(dict_info, ignore_index=True)


# 엑셀 내보내기
df.to_excel("쇼핑몰데이터.xlsx")

df
```
